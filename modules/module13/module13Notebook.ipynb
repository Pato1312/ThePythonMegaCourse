Not all datasets will have a header row (the one with titles) and if you call them it will look like this:

````python
df6 = pandas.read_csv("supermarkets/supermarkets-no-header.txt")
df6 
````

To solve this, you have to specify that there is no header after the file direction

````python
df6 = pandas.read_csv("supermarkets/supermarkets-no-header.txt", header = None)
df6 
````

Next we assign colums names and index:

````python
df6.columns = ["ID", "Address", "City", "ZIP code", "Country", "Name", "Employees"]
df6
````

To set index we use the set index function and we specify the columm we want to use, BUT this will no modify the existing data frame but instead will create a new one

````python
df7 = df6.set_index("ID")
df7
````

To not have to create another dataframe the call must be like this, but be carefull of moving the index as it will be deleting columms before index

````python
df6.set_index("ID", inplace=True)
df6
````

````python
df7.set_index("Address", inplace=True)
df7
````

to avoid that it must be called drop=false

````python
df7.set_index("ZIP code", inplace=True, drop=False)
df7
````

````python
import pandas
from geopy.geocoders import ArcGIS

nom = ArcGIS()
n = nom.geocode("3995 23rd St, San Francisco, CA 94114")
print("longitude : ", n.longitude, " - latitude : ",  n.latitude)

````

````python
geoDf = pandas.read_csv("supermarkets/supermarkets.csv")
geoDf
````

````python
geoDf["Full Address"] = geoDf["Address"] + ", " + geoDf["City"] + ", " + geoDf["State"] + ", " + geoDf["Country"]
geoDf["Coordinates"] =geoDf["Full Address"].apply(nom.geocode)
geoDf["Latitude"] = geoDf["Coordinates"].apply(lambda x: x.latitude if x != None else None)
geoDf["Longitude"] = geoDf["Coordinates"].apply(lambda x: x.longitude if x != None else None)
geoDf = geoDf.drop(columns = ["Full Address"])
geoDf
````
